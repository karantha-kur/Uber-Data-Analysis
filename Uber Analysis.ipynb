```python
# now we can move to converting the flafile into the facts and description table according to model 
```

```python
df = df.drop_duplicates().reset_index(drop = True)
df['trip_id'] = df.index
df
```

```python
# Creating the datetime table
date_time_table = df[['tpep_pickup_datetime', 'tpep_dropoff_datetime']].reset_index(drop = True)
date_time_table
```

```python
# adding columns 
date_time_table['pickup_time'] = date_time_table['tpep_pickup_datetime'].dt.time
date_time_table['pickup_date'] = date_time_table['tpep_pickup_datetime'].dt.date
date_time_table['pickup_day'] = date_time_table['tpep_pickup_datetime'].dt.day
date_time_table['dropoff_time'] = date_time_table['tpep_dropoff_datetime'].dt.time
date_time_table['dropoff_date'] = date_time_table['tpep_dropoff_datetime'].dt.date
date_time_table['dropoff_day'] = date_time_table['tpep_dropoff_datetime'].dt.day
date_time_table['date_time_id'] = date_time_table.index
date_time_table = date_time_table[['date_time_id', 'tpep_pickup_datetime', 'pickup_time', 'pickup_date', 'pickup_day', 'tpep_dropoff_datetime', 
                                   'dropoff_time', 'dropoff_date', 'dropoff_day']]

date_time_table
```

```python
# creating the ratecode table 
ratecode_type = {
    1:"Standard rate",
    2:"JFK",
    3:"Newark",
    4:"Nassau or Westchester",
    5:"Negotiated fare",
    6:"Group ride"
}
ratecode_table = df[['RatecodeID']].reset_index(drop = True)
ratecode_table['ratecode_id'] = ratecode_table.index 
ratecode_table['ratecode_name'] = ratecode_table['RatecodeID'].map(ratecode_type)
ratecode_table = ratecode_table[['ratecode_id', 'RatecodeID', 'ratecode_name']]
ratecode_table
```

```python
# create table dropoff_loc
dropoff_loc = df[['dropoff_longitude', 'dropoff_latitude']].reset_index(drop = True)
dropoff_loc['dropoff_loc_id'] = dropoff_loc.index
dropoff_loc = dropoff_loc[['dropoff_loc_id', 'dropoff_longitude', 'dropoff_latitude']]
dropoff_loc
```

```python
# creating the pickip_loc table 

pickup_loc = df[['pickup_longitude', 'pickup_latitude']].reset_index(drop = True)
pickup_loc['pickup_loc_id'] = pickup_loc.index
pickup_loc = pickup_loc[['pickup_loc_id', 'pickup_longitude', 'pickup_latitude']]
pickup_loc
```

```python
# creating the payment_type table 

payment_types = {
    1: 'Credit Card', 
    2: 'Cash',
    3: 'No Charge', 
    4: 'Dispute', 
    5: 'Unknown', 
    6: 'Voided Trip'
}

payment_type_table = df[['payment_type']].reset_index(drop = True)
payment_type_table['payment_type_id'] = payment_type_table.index
payment_type_table['payment_type_name'] = payment_type_table['payment_type'].map(payment_types)
payment_type_table = payment_type_table[['payment_type_id', 'payment_type', 'payment_type_name']]
payment_type_table
```

```python
# creating a table for trip distance
trip_distance_table = df[['trip_distance']].reset_index(drop = True)
trip_distance_table['trip_distance_id'] = trip_distance_table.index
trip_distance_table = trip_distance_table[['trip_distance_id', 'trip_distance']]
trip_distance_table
```

```python
# creating the passengen_count table 
passenger_count_table = df[['passenger_count']].reset_index(drop = True)
passenger_count_table['passenger_count_id'] = passenger_count_table.index
passenger_count_table = passenger_count_table[['passenger_count_id', 'passenger_count']]
passenger_count_table
```

```python
# creating the fare_amount_table
fare_amount_table = df[['fare_amount', 'extra',	'mta_tax', 'tip_amount', 'tolls_amount', 
                        'improvement_surcharge', 'total_amount']].reset_index(drop = True)
fare_amount_table['fare_amount_id'] = fare_amount_table.index
fare_amount_table = fare_amount_table[['fare_amount_id', 'fare_amount', 'extra', 'mta_tax', 'tip_amount', 
                                       'tolls_amount', 'improvement_surcharge', 'total_amount']]
fare_amount_table
```

```python
# now we need to create the fact table by joining data from thee tables
```

```python
fact_table = df.merge(passenger_count_table, left_on = 'trip_id', right_on = 'passenger_count_id') \
             .merge(trip_distance_table, left_on = 'trip_id', right_on = 'trip_distance_id') \
             .merge(ratecode_table, left_on = 'trip_id', right_on = 'ratecode_id') \
             .merge(pickup_loc, left_on = 'trip_id', right_on = 'pickup_loc_id') \
             .merge(dropoff_loc, left_on = 'trip_id', right_on='dropoff_loc_id') \
             .merge(date_time_table, left_on = 'trip_id', right_on='date_time_id') \
             .merge(payment_type_table, left_on = 'trip_id', right_on = 'payment_type_id') \
             .merge(fare_amount_table, left_on = 'trip_id', right_on = 'fare_amount_id') \
             [['trip_id','VendorID', 'date_time_id', 'passenger_count_id',
               'trip_distance_id', 'ratecode_id', 'store_and_fwd_flag', 'pickup_loc_id', 'dropoff_loc_id',
               'payment_type_id', 'fare_amount_id']]
```

```python
# now we can export the data to csv
fact_table.to_csv('fact_table.csv', index = False, header = True)
fare_amount_table.to_csv('fare_amount_table.csv', index = False, header = True)
passenger_count_table.to_csv('passenger_count_table.csv', index = False, header = True)
trip_distance_table.to_csv('trip_distance_table.csv', index = False, header = True)
payment_type_table.to_csv('payment_type_table.csv', index = False, header = True)
pickup_loc.to_csv('pickup_loc.csv', index = False, header = True)
dropoff_loc.to_csv('dropoff_loc.csv', index = False, header = True)
ratecode_table.to_csv('ratecode_table.csv', index = False, header = True)
date_time_table.to_csv('date_time_table.csv', index = False, header = True)
```
